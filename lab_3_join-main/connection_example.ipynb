 # Połączenie z bazą danych przy pomocy sqlAlchemy

```python
import sqlalchemy
from sqlalchemy import create_engine,text

connector = "postgresql"
user = "postgres"
password = "9731"
host = "localhost"
port = "5432"
dbname = "dvdrental"
db_string = "postgresql://postgres:9731@localhost:5432/dvdrental"
db_string = f"{connector}://{user}:{password}@{host}:{port}/{dbname}"
db = create_engine(db_string)
connection_sqlalchemy = db.connect()
```

 # Połączenie z bazą danych przy pomocy psycopg2 i pandas

```python
import psycopg2 as pg
import pandas as pd

connection = pg.connect(host=host, port=port, dbname=dbname, user=user, password=password);
```

Zapytanie do bazy i wypisanie wyników

```python
df = pd.read_sql('select * from city',con=connection)
df
```
